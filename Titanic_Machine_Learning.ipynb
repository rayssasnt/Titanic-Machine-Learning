# Encoding


```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', dtype='int32')

encoder = encoder.fit(treino[['Embarked']])
```

```python
encoder.transform(treino[['Embarked']]).toarray()
```

```python
encoder_df = pd.DataFrame(encoder.transform(treino[['Embarked']]).toarray(),
              columns = encoder.get_feature_names_out())
```

```python
encoder_df.head(3)
```

```python
treino = pd.concat([treino, encoder_df],axis=1)
```

```python
treino.head(3)
```

```python
encoder_df = pd.DataFrame(encoder.transform(teste[['Embarked']]).toarray(),
              columns = encoder.get_feature_names_out())
```

```python
teste = pd.concat([teste, encoder_df],axis=1)
```

```python
treino = treino.drop(['Embarked'],axis=1)
teste = teste.drop(['Embarked'],axis=1)
```

```python
teste.head(2)
```

```python
treino.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = treino.drop(['PassengerId','Survived'], axis=1)
y = treino['Survived']
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
X_treino , X_teste , y_treino , y_teste = train_test_split(X,y , test_size=0.3, random_state=46)
```

# Dummy : Modelo base     

```python
print(f"Treinaremos com {len(X_treino)} dados")
print(f"Testaremos com {len(X_teste)} dados")
```

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy = DummyClassifier(strategy='stratified') # inicia o modelo , sem o parametro 'stratified',
# o modelo sempre vai prever a classe mais comum do dataset


dummy.fit(X_treino,y_treino) # faz o ajuste
```

```python
y_pred = dummy.predict(X_teste) # faz a previsão
acuracia = dummy.score(y_teste,y_pred)*100 # acuracia
print(f"Acuracia: {acuracia:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_teste, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_teste, y_pred))
```

# Algoritmo KNN.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_treino, y_treino)
```

```python
acuracia_2 = knn.score(X_teste,y_teste)*100
print(f"A acurácia foi de {acuracia_2:.2f}%")
```

# DecisionTreeClassifier

```python
teste
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

classificador_tree = DecisionTreeClassifier(max_depth=3)
classificador_tree.fit(X_treino, y_treino)
previsoes = classificador_tree.predict(X_teste)

acuracia_3 = accuracy_score(y_teste, previsoes) * 100
print(f"A acurácia do modelo foi de {acuracia_3:.2f}%")
```

# LinearSVC

```python
from sklearn.svm import LinearSVC

modelo_svc = LinearSVC()
modelo_svc.fit(X_treino, y_treino)
previsoes = modelo_svc.predict(X_teste)

acuracia_4 = accuracy_score(y_teste, previsoes) * 100
print(f"A acurácia foi de {acuracia_4:.2f}%")
```

# Logistic Regression

Padronizando os dados para a regressão logística

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_treino)
X_test_scaled = scaler.transform(X_teste)
```

```python
from sklearn.linear_model import LogisticRegression

modelo_lr = LogisticRegression()

modelo_lr.fit(X_train_scaled,y_treino)
y_pred = modelo_lr.predict(X_test_scaled)

acuracia_5 = accuracy_score(y_teste, y_pred) * 100
print(f"A acurácia foi de {acuracia_5:.2f}%")
```

# Predição

Fazendo a predição dos dados para enviar ao kaggle

```python
X_teste = teste.drop("PassengerId",axis=1)
```

```python
y_pred = classificador_tree.predict(X_teste)
```

```python
teste['Survived'] = y_pred
```
